```python
#13
# Ví dụ 2 mảng
a = [3, 9, 1, 4]
b = [2, 7, 4, 3, 2, 8]

result = []

# Dùng vòng lặp với chỉ số nhỏ nhất
min_len = min(len(a), len(b))
for i in range(min_len):
    result.append(a[i] + b[i])

# Thêm phần tử còn lại của a hoặc b
result.extend(a[min_len:])  # Nếu a còn dư
result.extend(b[min_len:])  # Nếu b còn dư

print("Mảng kết quả:", result)
```

```python
#14
import random

#Nhập mảng
m = int(input("Nhập số dòng m: "))
n = int(input("Nhập số cột n: "))

#Câu a. Tạo mảng a chứa các số nguyên ngẫu nhiên
a = [[random.randint(1,10) for _ in range(n)] for _ in range(m)]

print("\n Câu a: Mảng a:")
for row in a:
  print(row)

#Câu b. Xuất các phần tử thuộc dòng k
k = int(input("\n Câu b: Nhập dòng k: "))
if 0 <= k < m:
  print(f"\nPhần tử thuộc dòng {k}:", a[k])
else:
  print("Dòng không hợp lệ")

#Câu c. Xuất các phần tử thuộc cột k
k = int(input("\n Câu c: Nhập cột k: "))
if 0 <= k < n:
  print("Cột", k, ":")
  for i in range(m):
    print(a[i][k])
else:
  print("Cột không hợp lệ")

#Câu d. Tìm dòng có tổng lớn nhất 45
print("\nCâu d: Các dòng có tổng > 45:")
for i in range(m):
  if(sum(a[i]) > 45):
    print(f"Dòng {i}:",a[i],"Tổng =",sum(a[i]))

#Câu e. Tìm cột có tích nhỏ nhất
min_tich = None
cot_min = -1

for j in range(n):
    tich = 1
    for i in range(m):
        tich *= a[i][j]
    if min_tich is None or tich < min_tich:
        min_tich = tich
        cot_min = j
print("\nCâu e: Cột có tích nhỏ nhất:", cot_min, "Tích =", min_tich)

# Câu f. Phần tử thuộc dòng chẵn và cột lẻ
print("\nCâu f: Phần tử dòng chẵn, cột lẻ:")
for i in range(0, m, 2):        # dòng chẵn
    for j in range(1, n, 2):    # cột lẻ
        print(a[i][j], end=" ")
print()

# Câu g. TBC các phần tử chẵn thuộc dòng lẻ
tong = 0
dem = 0
for i in range(1, m, 2):  # dòng lẻ
    for j in range(n):
        if a[i][j] % 2 == 0:
            tong += a[i][j]
            dem += 1

if dem > 0:
    print("\nCâu g: TBC số chẵn ở dòng lẻ:", tong / dem)
else:
    print("\nKhông có số chẵn ở dòng lẻ")

# Câu h. TBC các phần tử thuộc biên
tong = 0
dem = 0
for i in range(m):
    for j in range(n):
        if i == 0 or i == m-1 or j == 0 or j == n-1:
            tong += a[i][j]
            dem += 1

print("\nCâu h: TBC phần tử biên:", tong / dem)

# Câu i. TBC tích các phần tử không thuộc biên
tich = 1
dem = 0
for i in range(1, m-1):
    for j in range(1, n-1):
        tich *= a[i][j]
        dem += 1

if dem > 0:
    print("\nCâu i: TBC tích phần tử không biên:", tich ** (1/dem))
else:
    print("\nKhông có phần tử không biên")
```

```python
#15
class SinhVien:
    def __init__(self, ma, ten, namsinh, dtb):
        self.ma = ma
        self.ten = ten
        self.namsinh = namsinh
        self.dtb = dtb

# c. đủ điều kiện lên lớp
def dem_len_lop(ds):
    return sum(1 for sv in ds if sv.dtb >= 5)

# d. đủ 20 tuổi
def sv_20_tuoi(ds):
    return [sv for sv in ds if 2026 - sv.namsinh >= 20]

# e. hệ DH
def dem_DH(ds):
    return sum(1 for sv in ds if sv.ma[2:4] == "DH")

# f. tên Lan
def dem_ten_lan(ds):
    return sum(1 for sv in ds if "Lan" in sv.ten)

# g. họ Phan
def dem_ho_phan(ds):
    return sum(1 for sv in ds if sv.ten.startswith("Phan"))
```
